```python
def temperatura_hora_estado(estado):
  data_estado = np.genfromtxt('temperatura_' + estado + '.csv', delimiter=',', skip_header=1)
  data_estado = np.transpose(data_estado)
  return data_estado[1]
```

```python
tmp_sp = temperatura_hora_estado('SP')
tmp_rj = temperatura_hora_estado('RJ')
tmp_mg = temperatura_hora_estado('MG')
tmp_es = temperatura_hora_estado('ES')
```

```python
plt.boxplot([tmp_sp, tmp_rj, tmp_mg, tmp_es])
```

```python
ax1 = plt.subplot(2, 2, 1)
ax1.set_ylabel('Temperatura em (°C)')
ax1.set_xlabel('Estado')
ax1.set_title('Temperatura por estado por hora')
ax1.boxplot([tmp_sp, tmp_rj, tmp_mg, tmp_es], labels=['SP','RJ','MG','ES'])
ax1.plot()
```

```python
tmp_pe = temperatura_hora_estado('PE')
ax1 = plt.subplot(1, 1, 1) #número de linhas, numero de colunas e posição 
ax1.set_ylabel('Temperatura em (°C)')
ax1.set_xlabel('Estado')
ax1.set_title('Temperatura por estado por hora')
ax1.boxplot([tmp_pe], labels=['PE'])
ax1.plot()
```

```python
vento_km = vento_ms * 3.6
sensacao_termica_hora = 33 + (10 * np.sqrt(vento_km )+10.45 - vento_km) *(temperatura_hora-33)/22
```

```python
sensacao_termica_hora
```

```python
plt.plot(hora, sensacao_termica_hora, label = 'Sensação')
plt.plot(hora, temperatura_hora, label = 'Temperatura')
plt.legend()
```

```python
X = temperatura_hora
Y = vento_ms
n = np.size(temperatura_hora)
```

```python
a = (n * np.sum(X*Y)-np.sum(X)*np.sum(Y))/(n*np.sum(X**2)-np.sum(X)**2)
```

```python
b = np.mean(Y) - a * np.mean(X)
```

```python
y = a*X + b
```

```python
#gráfico de disperção
plt.grid()
plt.ylabel('Velocidade do vento por segundo')
plt.xlabel('Temperatura (°C)')
plt.title('Vento pela temperatura')
plt.scatter(temperatura_hora, vento_ms)
plt.plot(temperatura_hora, y, color='r')
```
