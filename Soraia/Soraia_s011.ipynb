```python
import numpy as np #importando biblioteca
```

```python
data = np.genfromtxt('temperatura_PE.csv', delimiter=',', skip_header=1) #importando os dados
```

```python
data =  np.transpose(data) #transpondo os dados
```

```python
data #printando para melhor visualização
```

```python
#atribuir valores aos parâmetros dos dados
#seguindo os índices

hora = data[0]
temperatura_hora = data[1]
```

```python
#média

np.mean(temperatura_hora)
```

```python
#mediana

np.median(temperatura_hora)
```

```python
#desvio padrão

np.std(temperatura_hora)
```

```python
#quartis e min e máx
min = np.min(temperatura_hora)
q1 = np.quantile(temperatura_hora, 0.25)
q2 = np.quantile(temperatura_hora, 0.5)
q3 = np.quantile(temperatura_hora, 0.75)
max = np.max(temperatura_hora)

print(f"A temperatura mínima registrada foi: {min:.2f}°C.\n O primeiro quartil é igual a: {q1:.2f}°C.\n O segundo quartil é igual a: {q2:.2f}°C.\n O terceiro quartil é igual a: {q3:.2f}°C.\n A temperatura máxima registrada foi: {max:.2f}°C.\n")
```

```python
import matplotlib.pyplot as plt
```

```python
def temperatura_hora_estado(estado):
  data_estado = np.genfromtxt('temperatura_' + estado + '.csv', delimiter=',', skip_header=1)
  data_estado = np.transpose(data_estado)
  return data_estado[1]
```

```python
tmp_pe = temperatura_hora_estado('PE')
plt.boxplot([tmp_pe])
ax1 = plt.subplot(1, 1, 1)

font1 = {'family':'arial','color':'red','size':15}
font2 = {'family':'arial','color':'red','size':12}

plt.title('Temperatura por hora', fontdict = font1)
ax1.set_title('Temperatura por hora')

ax1.set_ylabel('Temperatura em (°C)')
plt.ylabel('Temperatura em (°C)', fontdict = font2)

ax1.set_xlabel('Pernambuco')
plt.xlabel('Pernambuco', fontdict = font2) 

```

```python
data_bonus = np.genfromtxt('Maria_qtdd_agua.csv', delimiter=',', skip_header=1) 
data_bonus
#os dados são a quantidade em mls de quanto "maria" bebe de agua ao longo do dia!
```

```python
data_bonus
```

```python

dia = data_bonus[0]
quantidade_de_agua_por_dia = data_bonus[1]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(dia, quantidade_de_agua_por_dia)


font1 = {'family':'arial','color':'red','size':20}
font2 = {'family':'arial','color':'red','size':15}

plt.xlabel('Horas', fontdict = font2)
plt.ylabel('Quantidade de água (ml)', fontdict = font2)
plt.title('Quantidade de água em 24h que Maria consome', fontdict = font1)
```
